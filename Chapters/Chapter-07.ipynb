```r
user.packages <- c("data.table", "rgdal", "rgeos", "trend")
lapply(user.packages, require, character.only=T); rm(user.packages)

if(.Platform$OS.type!="windows"){
  paths <- c("/Volumes/Climate Rasters/", "/Volumes/Analysis Storage", "~/GitHub/Report/")} else {
    paths <- c("D:/Climate Rasters/", "E:/Analysis Storage/", "C:/Users/Cloud/GitHub/Report/")}
setwd(paths[3]); options(scipen=10); set.seed(12345); setDTthreads(16)
rasterOptions(progress="text", maxmemory=1e+09, chunksize=1e+08)
```

```r
cold.trend <- matrix(NA, nrow=cells, ncol=2)

for(i in 1:cell.list){
  values <- fread(paste(paths[2], "Cold Wave Large/Cell ", i, ".csv", sep=""))
  values <- values[!is.na(values$Max.Day), ]
  values[values$Month==2 & values$Day==29] <- NA
  trends <- ts(na.omit(values$MaxDay), frequency=365)
  try(cold.trend[i,1] <- sea.sens.slope(trends), silent=TRUE)
  try(cold.trend[i,2] <- smk.test(trends)$p.value, silent=TRUE)
  cat(paste("Current Cell:", i, "\n",seq=""))}
fwrite(cold.trend, "./Min_Temp_Trend_Data.csv")
sink(); Sys.sleep(900)


heat.trend <- matrix(NA, nrow=cells, ncol=2)

for(i in 1:cell.list){
  values <- fread(paste(paths[2], "Heat Wave Large/Cell ", i, ".csv", sep=""))
  values <- values[!is.na(values$MaxDay), ]
  values[values$Month==2 & values$Day==29] <- NA
  trends <- ts(na.omit(values$Max.Day), frequency=365)
  try(heat.trend[i,1] <- sea.sens.slope(trends), silent=TRUE)
  try(heat.trend[i,2] <- smk.test(trends)$p.value, silent=TRUE)
  cat(paste("Current Cell:", i, "\n",seq=""))}
fwrite(heat.trend, "./Max_Temp_Trend_Data.csv")
sink(); Sys.sleep(900)
```
